```python
# Import the required libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def tatto_match(img1, img2, x):
    # Initiate SIFT detector
    sift = cv2.xfeatures2d.SIFT_create()

    # find the keypoints and descriptors with SIFT
    kp1, des1 = sift.detectAndCompute(img1,None)
    kp2, des2 = sift.detectAndCompute(img2,None)

    # FLANN parameters
    FLANN_INDEX_KDTREE = 0
    index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
    search_params = dict(checks=50)   # or pass empty dictionary

    flann = cv2.FlannBasedMatcher(index_params,search_params)

    #menemukan kecocokan dengan FLANN
    matches = flann.knnMatch(des1,des2,k=2)
    

    # gambar line berdasarkan kecocokan,
    matchesMask = [[0,0] for i in range(len(matches))]
    good = [] 
    # ratio test as per Lowe's paper
    for i,(m,n) in enumerate(matches):
        if m.distance < 0.7*n.distance:
            matchesMask[i]=[1,0]
            good.append([m])

    # yang cocok digambar dengan warna kuning
    draw_params = dict(matchColor = (0,255,255),
                       singlePointColor = (255,0,0),
                       matchesMask = matchesMask,
                       flags = 0)
    
    # nilai descriptor bukti
    desbukti = len(des1)
    
    #jumlah fitur yang cocok
    cocok =len(good)
    
    # Hasil matching
    img3 = cv2.drawMatchesKnn(img1,kp1,img2,kp2,matches,None,**draw_params)

    fig.add_subplot(5,2, x)
    plt.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))     
    plt.title("Kecocokan yang ditemukan pada tersangka "+str(x)+' sebanyak : '+ str(cocok)+'/'+str(desbukti))
```

```python
# semua file gambar dapat di download pada link berikut:
# https://drive.google.com/drive/folders/1mBhHhwNAZEMpEpTX7OrVMNqqrXhhK-ge?usp=sharing

fig = plt.figure(figsize=(15, 30))

# membaca semua image tersangka
x = 1
while x < 11:
    img1 = cv2.imread('gambar/tersangka'+str(x)+'.jpg')
    img2 = cv2.imread('gambar/bukti.jpg')
    
    #resize ukuran gambar
    suspect = cv2.resize(img1, (1080,1080), interpolation = cv2.INTER_LINEAR)
    proof = cv2.resize(img2, (540,540), interpolation = cv2.INTER_LINEAR)
    
    #ubah gambar kedalam grayscale
    suspect = cv2.cvtColor(suspect, cv2.COLOR_BGR2GRAY)
    proof = cv2.cvtColor(proof, cv2.COLOR_BGR2GRAY)
    
    # masuk ke fungsi feature match menggunakan FLANN
    tatto_match(proof,suspect, x)
    
    x+=1
```
